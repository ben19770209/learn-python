題目:<br>
1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。<br>
2.呈上題，將以上數列取出偶數。<br>
3.呈1題，將數列取出3的倍數。<br>

```
import numpy as np
```

```
np.__version__
```

```
#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
ary1 = np.arange(0, 20+1, 1)
ary1
```

```
#2.呈上題，將以上數列取出偶數。
ary2 = np.array([x for x in ary1 if x % 2 == 0])
ary2
```

```
#3.呈1題，將數列取出3的倍數。
ary3 = np.array([x for x in ary1 if x % 3 == 0])
ary3
```

```
print(ary1.ndim, ary1.shape)
print(ary2.ndim, ary2.shape)
print(ary3.ndim, ary3.shape)
```
