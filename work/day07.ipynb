題目:<br>
array1 = np.array([[10, 8], [3, 5]]) <br>
1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```
import numpy as np
```

```
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
array1 = np.array([
                    [10, 8]
                    , [3, 5]
                    ])

print(array1)
```

```
ary1_inv = np.linalg.inv(array1)

print(ary1_inv)
```

```
ary1_mat = array1 @ ary1_inv

print(ary1_mat)
```

```
#2. 運用上列array計算特徵值、特徵向量?

w, v = np.linalg.eig(array1)

print(w)

print(v)
```

```
#3. 運用上列array計算SVD?

u, s, vh = np.linalg.svd(array1)

print(u)

print(s)

print(vh)
```
